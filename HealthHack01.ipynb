```python
blank = np.zeros((512,512,113))
my_pt = (254,222,31)
r = my_pt[0]
c = my_pt[1]
z = my_pt[2]
print(r)
print(c)
print(z)
```

```python
my_top_ht = blank
for i in range (-10,10):
    print(r+i)
    for j in range (-10,10):
        for k in range(-3, 3):
            my_top_ht[r+i, c+j, z+k] = 1.0
```

```python
import scipy.ndimage.filters as filters
```

```python
smth = 200*filters.gaussian_filter(my_top_ht,[50,50,10], [0,0,0] )
```

```python
import matplotlib.pyplot as plt
plt.imshow(smth[:,:,31], vmin =smth.min(), vmax = smth.max(), origin = "upper")
clb = plt.colorbar()
plt.show()
```
